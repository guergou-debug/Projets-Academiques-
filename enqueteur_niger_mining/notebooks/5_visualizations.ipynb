```python
# ==============================================================================
# NOTEBOOK 5: GÉNÉRATION DES 10 FIGURES CLÉS DE L'ÉTUDE
# PROJET: Text Mining - L'Enquêteur du Niger
# ==============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from collections import Counter
import os

# Configuration esthétique globale
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 11

# Chemins des fichiers
INPUT_PATH = "../data/final/enqueteur_final_analysis.csv"
FIGURES_DIR = "../outputs/figures/"
os.makedirs(FIGURES_DIR, exist_ok=True)

# Chargement du dataset enrichi
df = pd.read_csv(INPUT_PATH)
df['date_utc'] = pd.to_datetime(df['date_utc'])
df['texte_tokenise'] = df['texte_tokenise'].fillna("")
df['hashtags'] = df['hashtags'].fillna("")

print(f"Données chargées pour la visualisation : {df.shape[0]} articles.")
```

```python
# ==============================================================================
# BLOC 1 : ANALYSE TEXTUELLE & MOTS-CLÉS (MIS À JOUR ET COHÉRENT)
# ==============================================================================

# --- Définition unique des mots à masquer (Stop-words personnalisés) ---
# Centralisé ici pour que toutes les figures l'utilisent de la même manière.
mots_a_masquer = {
    'plus', 'comme', 'ici', 'cliquez', 'lien', 'vidéo', 'partager', 'page', 
    'suivre', 'direct', 'enquêteur', 'cette', 'journal', 'afrique', 'tout',
    'faire', 'très', 'être', 'avoir', 'aussi', 'sans', 'sous', 'vers', 'leurs', 'où', 'niger', 
    'nigeriens', 'nigerienne', 'nigeriennes', 'nigerien', 'nigérien', 'nigerienne', 'nigeriennes', 'a',
    'mais', 'donc', 'tous', 'toutes', 'toute', 'toutefois', 'pas', 'moins', 'beaucoup', 
    'peu', 'aucun', 'aucune', 'aucuns', 'aucunes', 'certaines', 'certains', 'ne', 'ni', 
    'ou', 'et', 'or', 'car', 'ceux', 'si', 'quand', 'lorsque', 'puisque', 'parce que', 
    'ainsi', 'alors', 'celui', 'cela', 'ceci', 'ces', 'ceux-ci', 'celles-ci', 'ceux-là', 
    'celles-là', 'celui-ci', 'celle-ci', 'celui-là', 'celle-là', 'là', 'delà', 'au-delà', 
    'en-deçà', 'dès', "jusqu'à", 'avant', 'après', 'pendant', 'depuis', 'contre', 'chez', 
    'entre', 'parmi', 'selon', 'malgré', 'grâce à','2025', '2026', '?'
}

# --- Figure 1 : Nuage de mots global ---
print("Génération de la Figure 1 (avec masquage des mots parasites)...")
texte_global = " ".join(df['texte_tokenise'])

wordcloud = WordCloud(
    width=1600, 
    height=800, 
    background_color='white', 
    colormap='viridis', 
    max_words=150,
    stopwords=mots_a_masquer
).generate(texte_global)

plt.figure(figsize=(16, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Nuage de mots global des articles de L'Enquêteur depuis l'investiture du 26 Mars 2025 à aujourd'hui", fontsize=20, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig(os.path.join(FIGURES_DIR, "fig1_nuage_mots_global2.png"), dpi=300)
plt.close()


# --- Figure 2 : Top 20 des mots les plus fréquents (Filtrée !) ---
print("Génération de la Figure 2 (avec le même filtre de mots masqués)...")

# On découpe le texte en mots individuels
tous_les_mots = texte_global.split()

# Filtrage : On ne garde QUE les mots qui ne sont PAS dans notre ensemble 'mots_a_masquer'
mots_filtres = [mot for mot in tous_les_mots if mot.lower() not in mots_a_masquer]

# On compte uniquement les mots restants
compteur_mots = Counter(mots_filtres)
mots_communs = pd.DataFrame(compteur_mots.most_common(20), columns=['Mot', 'Fréquence'])

plt.figure(figsize=(12, 6))
sns.barplot(data=mots_communs, x='Fréquence', y='Mot', palette='crest')
plt.title("Top 20 des mots significatifs les plus fréquents dans la ligne éditoriale", fontsize=14, fontweight='bold')
plt.xlabel("Nombre d'occurrences")
plt.ylabel("Mots")
plt.tight_layout()
plt.savefig(os.path.join(FIGURES_DIR, "fig2_top_20_mots.png"))
plt.close()


# --- Figure 3 : Top 15 des Hashtags / Slogans les plus utilisés (Nettoyage de sécurité) ---
print("Génération de la Figure 3...")

tous_les_hashtags = " ".join(df['hashtags']).split()

# On applique également le filtre sur les hashtags (au cas où des termes comme #niger y soient)
hashtags_filtres = [
    ht for ht in tous_les_hashtags 
    if ht.lower() not in mots_a_masquer and ht.replace('#', '').lower() not in mots_a_masquer
]

compteur_hashtags = Counter(hashtags_filtres)
hashtags_communs = pd.DataFrame(
    [item for item in compteur_hashtags.most_common(15) if item[0] != ""], 
    columns=['Hashtag', 'Fréquence']
)

plt.figure(figsize=(12, 6))
sns.barplot(data=hashtags_communs, x='Fréquence', y='Hashtag', palette='magma')
plt.title("Top 15 des slogans et hashtags de ralliement sur Facebook", fontsize=14, fontweight='bold')
plt.xlabel("Nombre d'utilisations")
plt.ylabel("Hashtags")
plt.tight_layout()
plt.savefig(os.path.join(FIGURES_DIR, "fig3_top_hashtags.png"))
plt.close()

print("Bloc 1 mis à jour avec succès et parfaitement harmonisé !")
```

```python
# ==============================================================================
# BLOC 2 : ANALYSE DE SENTIMENT & INTENSITÉ
# ==============================================================================

# --- Figure 4 : Distribution de la Polarité (Le ton général) ---
print("Génération de la Figure 4...")
plt.figure(figsize=(10, 5))
sns.histplot(df['polarite'], kde=True, color='crimson', bins=30)
plt.axvline(x=0, color='black', linestyle='--', alpha=0.7, label='Neutralité (0.0)')
plt.axvline(x=df['polarite'].mean(), color='blue', linestyle='-', linewidth=2, label=f'Moyenne ({df["polarite"].mean():.2f})')
plt.title("Distribution de la polarité des articles (Ton du journal)", fontsize=14, fontweight='bold')
plt.xlabel("Polarité (Négatif < 0 | Positif > 0)")
plt.ylabel("Nombre d'articles")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(FIGURES_DIR, "fig4_distribution_polarite.png"))
plt.close()


# --- Figure 5 : Relation entre Polarité et Subjectivité ---
print("Génération de la Figure 5 (avec des noms de thèmes explicites)...")

# 1. Définis ici la correspondance entre tes numéros de thèmes et leurs vrais noms.
# (Ajuste les noms à droite en fonction de la nomenclature exacte de ton étude !)
dictionnaire_themes = {
    0: "Thème 0 (ex: Économie / Développement)",
    1: "Souveraineté & Coopération AES",
    2: "Politique Interne & Transition",
    3: "Sécurité & Opérations Militaires",
    4: "Thème 4 (ex: Médias & Société)"
    # Ajoute ou modifie les clés (qui peuvent être des int ou des str, ex: '1' au lieu de 1)
}

# 2. Copie temporaire pour ne pas altérer le DataFrame original
df_visu = df.copy()

# 3. Conversion propre de la colonne pour appliquer le dictionnaire
# On s'assure que les types correspondent (ici on force en entier si tes thèmes sont numériques)
try:
    df_visu['theme_dominant_nom'] = df_visu['theme_dominant'].astype(int).map(dictionnaire_themes)
except ValueError:
    # Si tes thèmes sont déjà des chaînes de caractères (ex: "1", "2")
    df_visu['theme_dominant_nom'] = df_visu['theme_dominant'].astype(str).map(dictionnaire_themes)

# Si certains thèmes n'ont pas de correspondance, on garde leur valeur d'origine
df_visu['theme_dominant_nom'] = df_visu['theme_dominant_nom'].fillna(df_visu['theme_dominant'])

# 4. Tracé du graphique avec la nouvelle colonne explicite
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_visu, 
    x='polarite', 
    y='subjectivite', 
    alpha=0.7, 
    hue='theme_dominant_nom',  # <-- On utilise la colonne avec les vrais noms !
    palette='Set1'
)

# Ligne verticale de neutralité
plt.axvline(x=0, color='gray', linestyle='--', alpha=0.5)

# Titres et labels
plt.title("Positionnement sémantique : Polarité vs Subjectivité", fontsize=14, fontweight='bold')
plt.xlabel("Polarité (Négatif vs Positif)")
plt.ylabel("Subjectivité (Faits vs Opinion)")

# Positionnement de la légende pour qu'elle soit bien lisible et ne cache pas les points
plt.legend(title="Thématique dominante", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.savefig(os.path.join(FIGURES_DIR, "fig5_polarite_vs_subjectivite.png"), bbox_inches='tight')
plt.close()

print("Figure 5 corrigée avec succès !")
```

```python
# ==============================================================================
# BLOC 3 : DYNAMIQUE TEMPORELLE
# ==============================================================================

# --- Figure 6 : Évolution de l'indice de sentiment au cours du temps ---
print("Génération de la Figure 6...")
# Moyenne glissante sur 15 jours de la polarité pour lisser les courbes temporelles
df_temp = df.sort_values(by='date_utc')
df_temp.set_index('date_utc', inplace=True)
polarite_glissante = df_temp['polarite'].rolling('15D', min_periods=2).mean()

plt.figure(figsize=(14, 6))
polarite_glissante.plot(color='darkred', linewidth=2)
plt.axhline(0, color='black', linestyle='--', alpha=0.5)
plt.title("Évolution temporelle de la polarité (Moyenne glissante 15 jours)", fontsize=14, fontweight='bold')
plt.xlabel("Date de publication")
plt.ylabel("Polarité moyenne")
plt.tight_layout()
plt.savefig(os.path.join(FIGURES_DIR, "fig6_evolution_sentiment_temps.png"))
plt.close()
```

```python
# ==============================================================================
# BLOC 4 : CROISEMENT AVEC L'ENGAGEMENT (MÉTADONNÉES)
# ==============================================================================

# --- Figure 7 : Polarité de l'article vs Taux de partage ---
print("Génération de la Figure 7...")
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='polarite', y='nb_partages', scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title("Est-ce que la virulence d'un post (polarité négative) augmente ses partages ?", fontsize=14, fontweight='bold')
plt.xlabel("Polarité du texte")
plt.ylabel("Nombre de partages")
plt.tight_layout()
plt.savefig(os.path.join(FIGURES_DIR, "fig7_polarite_vs_partages.png"))
plt.close()


# --- Figure 8 : Profil des réactions sur Facebook face aux articles ---
print("Génération de la Figure 8...")
reactions_totales = df[['like', 'love', 'solidaire', 'haha', 'wow', 'triste', 'colere']].sum()
reactions_totales_df = pd.DataFrame({'Réaction': reactions_totales.index, 'Total': reactions_totales.values})

plt.figure(figsize=(10, 5))
sns.barplot(data=reactions_totales_df, x='Total', y='Réaction', palette='viridis')
plt.title("Distribution cumulative des types de réactions de l'audience", fontsize=14, fontweight='bold')
plt.xlabel("Nombre total de réactions enregistrées")
plt.ylabel("Réactions Facebook")
plt.xscale('log')  # Échelle logarithmique car les 'likes' écrasent souvent tout le reste
plt.tight_layout()
plt.savefig(os.path.join(FIGURES_DIR, "fig8_distribution_reactions.png"))
plt.close()
```

```python
# ==============================================================================
# BLOC 5 : THÉMATIQUES & SENTIMENTS (MIS À JOUR - AXES EXPLICITES)
# ==============================================================================

print("Préparation des données thématiques explicites pour le Bloc 5...")

# 1. Utilisation du même dictionnaire de correspondance que la Figure 5
# (Ajuste les noms à droite selon ta nomenclature exacte)
dictionnaire_themes = {
    0: "Économie & Développement",
    1: "Souveraineté & AES",
    2: "Politique & Transition",
    3: "Sécurité & Militaire",
    4: "Médias & Société"
}

# 2. Copie et transformation propre des identifiants de thèmes en labels explicites
df_bloc5 = df.copy()

try:
    df_bloc5['theme_label'] = df_bloc5['theme_dominant'].astype(int).map(dictionnaire_themes)
except ValueError:
    df_bloc5['theme_label'] = df_bloc5['theme_dominant'].astype(str).map(dictionnaire_themes)

# Sécurité : si un ID n'est pas dans le dictionnaire, on garde sa valeur d'origine
df_bloc5['theme_label'] = df_bloc5['theme_label'].fillna(df_bloc5['theme_dominant'])


# --- Figure 9 : Nombre d'articles et polarité moyenne par Thème ---
print("Génération de la Figure 9...")
fig, ax1 = plt.subplots(figsize=(12, 6))

# Calcul des volumes et moyennes groupés par notre nouveau label
stats_par_theme = df_bloc5.groupby('theme_label').agg(
    volume=('theme_label', 'count'),
    polarite_moyenne=('polarite', 'mean')
).sort_index()

# Axe 1 : Histogramme du volume d'articles (barres bleues)
sns.barplot(
    x=stats_par_theme.index, 
    y=stats_par_theme['volume'], 
    alpha=0.6, 
    color='royalblue', 
    ax=ax1
)
ax1.set_ylabel("Nombre d'articles (Volume)", color='royalblue', fontweight='bold')
ax1.tick_params(axis='y', labelcolor='royalblue')
ax1.set_xlabel("Thématique dominante", fontweight='bold', labelpad=10)

# Axe 2 : Ligne de la polarité moyenne par thème (courbe rouge)
ax2 = ax1.twinx()
sns.lineplot(
    x=stats_par_theme.index, 
    y=stats_par_theme['polarite_moyenne'], 
    marker='o', 
    color='crimson', 
    linewidth=3, 
    markersize=8,
    ax=ax2
)
ax2.set_ylabel("Polarité moyenne (Sentiment)", color='crimson', fontweight='bold')
ax2.tick_params(axis='y', labelcolor='crimson')
ax2.axhline(0, color='black', linestyle='--', alpha=0.5)  # Seuil de neutralité

# Rotation des labels sur l'axe X pour éviter les chevauchements de longs textes
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=15, ha='right')

plt.title("Volume de publications et polarité par Thématique", fontsize=14, fontweight='bold', pad=15)
plt.tight_layout()
plt.savefig(os.path.join(FIGURES_DIR, "fig9_theme_volume_polarite.png"), bbox_inches='tight')
plt.close()


# --- Figure 10 : Taux d'engagement moyen par Thème ---
print("Génération de la Figure 10...")

# Calcul de l'engagement total (likes + partages + commentaires)
df_bloc5['engagement_total'] = df_bloc5['like'] + df_bloc5['nb_partages'] + df_bloc5['nb_commentaires']

# Calcul de la moyenne par label explicite
theme_engagement = df_bloc5.groupby('theme_label')['engagement_total'].mean().reset_index().sort_values(by='theme_label')

plt.figure(figsize=(12, 6))
sns.barplot(
    data=theme_engagement, 
    x='theme_label', 
    y='engagement_total', 
    palette='flare'
)

plt.title("Niveau d'engagement moyen des lecteurs par Thématique", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Thématique dominante", fontweight='bold', labelpad=10)
plt.ylabel("Moyenne d'interactions (Likes + Partages + Comm.)", fontweight='bold')

# Rotation des étiquettes de l'axe X
plt.xticks(rotation=15, ha='right')

plt.tight_layout()
plt.savefig(os.path.join(FIGURES_DIR, "fig10_engagement_moyen_par_theme.png"), bbox_inches='tight')
plt.close()

print("\nFÉLICITATIONS ! Vos 10 figures stratégiques sont désormais harmonisées, lisibles et prêtes pour votre rapport.")
```
